```python
#exercise1 to jest dobrze
import pandas as pd
import json

# Wczytanie pliku csv do dataframe
df = pd.read_csv('proj1_ex01.csv')

# inicjalizacja pustej list w której bedziemy przechowywać informacje o kolumnach
columns_info = []

# iterujemy po wszystkich kolumnach w tabeli 
for col in df.columns:
    # Dla każdej kolumny, ta linia kodu oblicza procentowy udział brakujących wartości. df[column] wybiera kolumnę o nazwie column, a metoda .isnull() zwraca DataFrame, w którym każda wartość jest wartością logiczną (True jeśli brakuje wartości, False w przeciwnym razie). Metoda .mean() oblicza średnią wartość logiczną, co daje procentowy udział brakujących wartości w danej kolumnie.
    missing_percentage = df[col].isnull().mean()

    data_type = 'other'
    if df[col].dtype == 'int64':
        data_type = 'int'
    elif df[col].dtype == 'float64':
        data_type = 'float'

    #  Tworzy słownik column_dict zawierający informacje o kolumnie, takie jak nazwa, procent brakujących wartości i typ danych.
    col_info = {
        'name': col,
        'missing': round(missing_percentage, 2),
        'type': data_type
    }

    # dodajemy info o kolumnach do listy 
    columns_info.append(col_info)

#  Ten blok kodu otwiera plik o nazwie 'proj1_ex01_fields.json' w trybie zapisu ('w').
with open('proj1_ex01_fields.json', 'w') as f:
    json.dump(columns_info, f, indent=4)

```

```python
#exercise 3
import pandas as pd

# Wczytaj plik CSV do DataFrame
df = pd.read_csv('proj1_ex01.csv')

# Zmień nazwy kolumn zgodnie z określonymi zasadami
nowe_kolumny = {}
for kolumna in df.columns:
    nowa_kolumna = ''.join(znak.lower() if znak.isalnum() or znak == '_' else '_' for znak in kolumna)
    nowa_kolumna = nowa_kolumna.replace('some;name', 'somename')  # Zmiana 'some_name' na 'somename'
    nowe_kolumny[kolumna] = nowa_kolumna

# Popraw nazwy kolumn zgodnie z informacją zwrotną
nowe_kolumny['What is this even?'] = 'what_is_this_even'
df.rename(columns=nowe_kolumny, inplace=True)

# Zapisz DataFrame z nowymi nazwami kolumn do pliku CSV
df.to_csv('proj1_ex03_columns.csv', index=False)

```

```python
#exercise 5
import pandas as pd

# Załaduj DataFrame z pliku pickle o nazwie proj1_ex05.pkl
df = pd.read_pickle('proj1_ex05.pkl')

# Wybierz 2. i 3. kolumnę niezależnie od ich nazw
wybrane_kolumny = df.iloc[:, 1:3]

# Wybierz wiersze, których indeksy zaczynają się od litery 'v'
wybrane_wiersze = df[df.index.str.startswith('v')]

# Połącz wybrane kolumny i wiersze
wybrane_dane = wybrane_kolumny.loc[wybrane_wiersze.index]

# Zamień brakujące wartości na pusty ciąg znaków
wybrane_dane = wybrane_dane.fillna('')

# Utwórz tabelę Markdown z wybranych danych
tabela_markdown = wybrane_dane.to_markdown()

# Zapisz tabelę Markdown do pliku
with open('proj1_ex05_table.md', 'w') as plik:
    plik.write(tabela_markdown)



```
