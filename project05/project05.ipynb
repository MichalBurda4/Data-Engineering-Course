```python
#Import bibliotek
import pandas as pd
import numpy as np
import json
```

```python
#The file proj5_params.json contains a JSON dictionary with parameters which will be used
#throughout the exercises. Load it.
```

```python
params = {}
with open('proj5_params.json') as f:
    params = json.load(f)
print(params)
```

```python
#Exercise 1: Preparing time series datasets
```

```python
#The file proj5_timeseries.csv contains time series data. The first column contains timestamps,
#and the subsequent columns contain cumulative values for various parameters.
```

```python
#Perform the usual cleaning of column names – make letters (A–Z) lowercase and replace any other
#characters with underscores.
```

```python
df = pd.read_csv("proj5_timeseries.csv")
df.columns = df.columns.str.lower().str.replace(r"[^a-z]", "_", regex=True)
```

```python
df
```

```python
#Convert the values in the first column to pandas datetime objects and set the column as the index.

# Konwersja wartości w pierwszej kolumnie na obiekty datetime i ustawienie tej kolumny jako indeks
df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0], format='mixed', errors='coerce')
df.set_index(df.columns[0], inplace=True)
```

```python
#Save the cleaned DataFrame to proj5_ex01.pkl
#The original_frequency contains the appropriate frequency symbol for the test dataset. Set the
#appropriate frequency in your DataFrame’s index.
df = df.asfreq(params['original_frequency'])
```

```python
df.to_pickle("proj5_ex01.pkl")
df_copy = df.copy()
```

```python
df_copy = df.copy()
```

```python
df
```

```python
#Exercise 2 
```

```python
#Change your DataFrame’s frequency to the one indicated by the target_frequency parameter.
```

```python
df = df_copy.copy()
df = df.asfreq(params['target_frequency'])
```

```python
#Save the resulting DataFrame to proj5_ex02.pkl.
df.to_pickle("proj5_ex02.pkl")
```

```python
#Exercise 3 
```

```python
#Downsample your DataFrame accordingly, using an aggregation function appropriate for cumulative
#data, and make sure that the aggregates for a given “window” are calculated only if all samples
#from that window are present (otherwise, just put NaN in).
```

```python
#Parameters downsample_periods and
#downsample_units contain the number of periods (e.g. 3) and the units in which the periods
#are expressed (e.g. d for days), respectively.
```

```python
df = df_copy.copy()
df = df.resample(str(params['downsample_periods']) + params['downsample_units']).sum(min_count=params['downsample_periods'])
```

```python
df
```

```python
#Save the resulting DataFrame to proj5_ex03.pkl.
df.to_pickle("proj5_ex03.pkl")
```

```python
#Exercise 4 
```

```python
#Parameters upsample_periods and upsample_units contain the number of periods (e.g. 2) and
#the units in which the periods are expressed (e.g. h for hours), respectively.
#Parameters interpolation and interpolation_order contain the interpolation type
#(e.g. polynomial) and order (e.g. 3).
```

```python
df = df_copy.copy()
df = df.resample(str(params['upsample_periods']) + params['upsample_units']).interpolate(params['interpolation'], order=params['interpolation_order'])
```

```python
df
```

```python
#Upsample your original DataFrame to the target frequency and perform appropriate interpolation.
```

```python
ratio = pd.Timedelta(params['upsample_periods'], params['upsample_units']) / pd.Timedelta(1, params['original_frequency'])
df *= ratio
```

```python
#Save the DataFrame to proj5_ex04.pkl.
df.to_pickle("proj5_ex04.pkl")
```

```python
#Read the file proj5_sensors.csv. The file contains a date/time index and two columns:
```

```python
df = pd.read_pickle("proj5_sensors.pkl")
```

```python
df
```

```python
#The file contains a date/time index and two columns:
#• device_id, containing the device identifier (this can be a number or a string),
#• value, containing the reading value.
```

```python
df = df.pivot(columns='device_id', values='value')
```

```python
#Parameters sensors_periods and sensors_units contain the number of periods (e.g. 3) and the
#units in which the periods are expressed (e.g. d for days), respectively.
```

```python
new_index = pd.date_range(start=df.index.round("1min").min(), end=df.index.round("1min").max(), freq=str(params['sensors_periods']) + str(params['sensors_units']))
```

```python
#in case a value for a specific timestamp isn’t available, the values should be interpolated using
#linear interpolation,
```

```python
df = df.reindex(new_index.union(df.index)).interpolate()
df = df.reindex(new_index)
df = df.dropna()
```

```python
#Save the new DataFrame to proj5_ex05.pkl.
df.to_pickle("proj5_ex05.pkl")
```
