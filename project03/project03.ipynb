```python
#2.1
#There are three homogenous JSON files called:
#• proj3_data1.json,
#• proj3_data2.json,
#• proj3_data3.json.
#Load all these files into one big DataFrame, with an index containing a consistent sequence of
#numbers, e.g. 0, 1, 2, …. Keep the original column structure.
#Save the DataFrame to proj3_ex01_all_data.json. Use the default orient used by to_json(),
#i.e. columns.
```

```python
import pandas as pd

files_list = ['proj3_data1.json', 'proj3_data2.json', 'proj3_data3.json']
dfs = [] #lista, która będzie przechowywać poszczególne ramki danych


for file in files_list:
    dfs.append(pd.read_json(file))
    
df = pd.concat(dfs, ignore_index=True)
print(df.to_string())

df.to_json('proj3_ex01_all_data.json') #zapis do pliku
```

```python
#2.2 Exercise 2: Missing values
#2 points
#Produce a CSV file containing column names and the number of missing values in these columns.
#Only include the columns which have any missing values, e.g.:
#doors,2
#fuel_consumption,4
#Save the file as proj3_ex02_no_nulls.csv.
```

```python
import csv
```

```python
import csv

brakujace_kolumny = []

# Iterujemy przez kolumny DataFrame'a
for kolumna in df.columns:
    # Obliczamy liczbę brakujących wartości
    liczba_brakujacych_wartosci = df[kolumna].isnull().sum()
    if liczba_brakujacych_wartosci > 0:
        # Dodajemy krotkę (kolumna, liczba brakujących wartości) do listy
        brakujace_kolumny.append((kolumna, liczba_brakujacych_wartosci))

# Tworzymy plik CSV zawierający nazwy kolumn i liczbę brakujących wartości w tych kolumnach
# Zapisujemy plik jako proj3_ex02_no_nulls.csv
with open('proj3_ex02_no_nulls.csv', mode='w') as plik_csv:
    writer = csv.writer(plik_csv)
    
    # Zapisujemy każdy wiersz (kolumna, liczba brakujących wartości) do pliku CSV
    for wiersz in brakujace_kolumny:
        writer.writerow(wiersz)
```

```python
#2.3 Exercise 3: Applying functions
#2 points
#The file proj3_params.json contains a dictionary with certain parameters which will be important
#in the following exercises. Load the file so that it is available later on.
#Add a column called description to the DataFrame you created in the Exercise 1. It should
#contain the values from columns listed as the concat_columns parameter, separated with spaces.
#E.g., if concat_columns is ['make', 'model', 'engine'], for the following record:
#• make: Audi,
#• model: A3,
#• engine: 1.5 TFSI,
#the description column should contain the string “Audi A3 1.5 TFSI”.
#Save the DataFrame as proj3_ex03_descriptions.json. Again, use the default orient.
```

```python
import json
with open('proj3_params.json') as json_file:
    data = json.load(json_file)
print(data)
df['description'] = df[data['concat_columns']].apply(lambda x: ' '.join(x), axis=1)
print(df.to_string())
df.to_json('proj3_ex03_descriptions.json')
```

```python
df_json = pd.read_json('proj3_more_data.json')
list_of_columns_to_join_on = data['join_column']
print(list_of_columns_to_join_on)

df = df.merge(df_json, on=list_of_columns_to_join_on, how='left')
print(df.to_string())
df.to_json('proj3_ex04_joined.json')
```

```python
#2.5 Exercise 5: Iterating over DataFrames
#4 points
#Create separate JSON files for each record in your DataFrame obtained in Exercise 4. The JSON
#record should include all fields apart from description.
#The file name for each should be based on the value of description and follow the
#proj3_ex05_(description).json pattern. The value of description should be converted to
#lowercase, with spaces replaced with underscores.
#For example, for the record with “Audi A3 1.5 TFSI” as the description, the file name should be
#proj3_ex05_audi_a3_1.5_tfsi.json.
#Now repeat the following task, but make sure that columns listed in the int_columns parameter
#are saved as integers, not floats. Also make sure that the JSON file does not contain NaN values,
#as they are not allowed by the specification (only null is, and that should be used).
#Save the files in a similar way as before, but now use the proj3_ex05_int_(description).json
#pattern; e.g., for the example above, the file name should be
#proj3_ex05_int_audi_a3_1.5_tfsi.json.
```

```python
for index, wiersz in df.iterrows():
    opis = wiersz['description'].lower().replace(' ', '_')  # Przekształć opis na małe litery i zamień spacje na podkreślenia
    wiersz = wiersz.drop(['description']).to_dict()  # Usuń kolumnę 'description' i zamień wiersz na słownik

    # Zamień wszystkie brakujące wartości na None
    for klucz, wartosc in wiersz.items():
        if pd.isna(wartosc):
            wiersz[klucz] = None
    
    # Zapisz do pliku JSON
    nazwa_pliku = f"proj3_ex05_{opis}.json"
    with open(nazwa_pliku, 'w') as plik_json:
        json.dump(wiersz, plik_json)
    
    kolumny_liczbowe = data['int_columns']  # Załóżmy, że int_columns to lista nazw kolumn z wartościami liczbowymi
    for kolumna in kolumny_liczbowe:
        if pd.notnull(wiersz[kolumna]):
            wiersz[kolumna] = int(wiersz[kolumna])  # Jeśli wartość nie jest brakująca, zamień na liczbę całkowitą
        else:
            wiersz[kolumna] = None
    
    # Ponownie zamień brakujące wartości na None
    for klucz, wartosc in wiersz.items():
        if pd.isna(wartosc):
            wiersz[klucz] = None
    
    # Zapisz do pliku JSON z oznaczeniem, że są to wartości liczbowe
    nazwa_pliku_liczbowego = f"proj3_ex05_int_{opis}.json"
    with open(nazwa_pliku_liczbowego, 'w') as plik_json:
        json.dump(wiersz, plik_json)
```

```python
#2.6 Exercise 6: Aggregation
#4 points
#The aggregations parameter contains the aggregations which should be performed for the data,
#in the form of tuples, where the first element is the column name and the second element is the
#aggregation function, e.g.:
#('displacement', 'min'),
#('displacement', 'max'),
#('fuel_consumption', 'mean')
#Create a JSON file called proj3_ex06_aggregations.json which contains the aggregated values
#from DataFrame obtained in Exercise 4, in the form of a dictionary, where the keys are composed
#from the aggregation function and the column name, joined with an underscore, e.g.:
#{
#"min_displacement": 875.0,
#"max_displacement": 3000.0,
#"mean_fuel_consumption": 6.414285714285714
#}
```

```python
import numpy as np

aggregations_list = data['aggregations']
print(aggregations_list)

aggregations_dict = {}

for col, agg_func in aggregations_list:
    key = f"{agg_func}_{col}"
    # apply aggregation function to the columns
    aggregations_dict[key] = df[col].agg(agg_func)
print(aggregations_dict)

with open('proj3_ex06_aggregations.json', 'w') as f:
    json.dump(aggregations_dict, f)
```

```python
#2.7 Exercise 7: Grouping
#2 points
#Group the values from DataFrame obtained in the Exercise 4 by the column indicated as the
#grouping_column parameter. Create a DataFrame which contains the mean values, for all groups
#and all numerical columns, but only for groups containing more than 1 record.
#Save the result as proj3_ex07_groups.csv, including the header and the index
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ex04df = pd.read_json("proj3_more_data.json")

merged_df = pd.merge(df, ex04df, on=data["join_column"], how="left")
groups = merged_df.groupby(data['grouping_column']).filter(lambda x: len(x) > 1)

df_7 = pd.DataFrame()
for i,g in groups.groupby(data['grouping_column']):
       df_7[i] = g.mean(numeric_only=True)

df_7 = df_7.T
df_7.index.name = data['grouping_column']
df_7.to_csv('proj3_ex07_groups.csv', header=True, index=True)
df_7
```

```python
reshape1_df = df.groupby([data["pivot_index"], data["pivot_columns"]])[data["pivot_values"]].max()
# pivot to desired shape
reshape1_df = reshape1_df.reset_index().pivot(index=data["pivot_index"], columns=data["pivot_columns"], values=data["pivot_values"])
# remove rows with all nulls
reshape1_df = reshape1_df[reshape1_df.notnull().any(axis=1)]
print(reshape1_df.to_string())

# save to json
reshape1_df.to_pickle('proj3_ex08_pivot.pkl')
```
